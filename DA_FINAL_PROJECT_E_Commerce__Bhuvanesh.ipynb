# Project Title


Exploratory Data Analysis and Customer Churn Modeling for E-Commerce Customer Insights Using Python

# Domain

 E-commerce

# Objective

1. To explore and understand the e-commerce dataset using Exploratory Data Analysis (EDA).

2. To analyze customer behavior and purchase patterns such as product categories, order value, and country distribution.

3. To identify factors that influence customer churn based on customer activity and subscription status.

4. To provide actionable insights and recommendations from the analysis

# Outcome

**To identify the key factors influencing customer churn and predict whether a customer is likely to remain active or leave the e-commerce platform.**


#Dataset Information

Source: Source: Kaggle – E-Commerce Customer Insights and Churn Dataset

Year / Timeline: Data collected during 2021 to 2025

Dataset Description:This dataset contains 2,000 customer-level and order-level records from a simulated e-commerce platform. It includes information related to customer demographics, purchasing behavior, order history, and churn status.

Order Id - Unique identifier for each order placed on the platform.

Customer Id - Unique identifier assigned to each customer.

Customer AGe - Age of the customer who placed the order.

product_id - Unique identifier for the purchased product.

country - Country where the customer is located.

signup_date - Date when the customer registered on the platform.

last_purchase_date - The most recent purchase date of the customer before the current order.

cancellations_count - Total number of orders cancelled by the customer.

subscription_status - Current subscription status of the customer (active, paused, cancelled).

order_date - Date when the order was placed.

unit_price - Price of a single unit of the product.

quantity - Number of units purchased in the order.

purchase_count - Total number of purchases made by the customer.

preferred_category - Product category most frequently purchased by the customer.

product_name - Name of the product purchased.

category - Product category (Electronics, Clothing, Sports, Beauty, etc.)

gender - Gender of the customer.

# Type of Analysis

Exploratory Data Analysis (EDA) is used to understand the dataset, identify patterns, and analyze customer behavior.

Predictive Analysis is used to build a model that predicts whether a customer is likely to churn or remain active.

## Stage 2 – Data Cleaning and Pre-processing


```python
# Clean column names
df.columns = df.columns.str.strip()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Handle duplicates
df.duplicated().sum()
```

```python
# Convert date columns
df["signup_date"] = pd.to_datetime(df["signup_date"])
df["last_purchase_date"] = pd.to_datetime(df["last_purchase_date"])
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
# Treat Outliers
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df["age"])
plt.show()

sns.boxplot(x=df["unit_price"])
plt.show()

sns.boxplot(x=df["quantity"])
plt.show()
```

```python
# Check skewness
df[["age","unit_price","quantity"]].skew()
```

```python
df["total_order_value"] = df["unit_price"] * df["quantity"]
```

```python
display(df)
```

```python
# Cleaned Dataset
df.to_csv("cleaned_dataset_E_commerce_.csv", index = False)
```

## Stage 3 – EDA and Visualizations



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1 Countplot – Orders by Country
plt.figure(figsize=(8,5))
sns.countplot(x="country", data=df)
plt.title("Orders by Country")
plt.xticks(rotation=45)
plt.show()
```

## Interpretation of **1.Orders by Country (Countplot)**

The above countplot visualizes the distribution of customer orders across different countries.

The chart compares how many orders were placed by customers from each country, showing which regions contribute more to the platform’s overall activity.

Features

Several key visual features are used in this chart:

X-axis (Country): Represents the different countries where customers are located.

Y-axis (Count): Represents the number of orders placed by customers.

Bars: Each bar shows the total number of orders from a particular country.

Insights

Some countries have significantly higher order counts than others.

This indicates stronger customer engagement in those regions.

Businesses can use this insight to focus marketing strategies on high-performing countries

```python
# 2.Countplot - Orders by Category
plt.figure(figsize=(8,5))
sns.countplot(x="category", data=df)
plt.title("Orders by Product Category")
plt.xticks(rotation=45)
plt.show()
```

## Interpretation of **Chart 2 – Orders by Product Category (Countplot)**

Interpretation of the above chart

The above countplot visualizes the number of orders placed across different product categories.

It helps compare which categories customers purchase most frequently.

Features

X-axis (Category): Represents different product categories available on the platform.

Y-axis (Count): Represents the total number of orders placed in each category.

Bars: Show the popularity of each product category.

Insights

Some product categories receive more orders than others.

High-demand categories can be prioritized for inventory and promotions.

```python
# 3 Histogram (Age Distribution)
plt.figure(figsize=(8,5))
plt.hist(df["age"], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Interpretation of **Chart 3 – Customer Age Distribution (Histogram)**

The above histogram visualizes the distribution of customers based on their age.

It helps understand which age groups are most active on the e-commerce platform.

Features

X-axis (Age): Represents customer age groups.

Y-axis (Frequency): Represents the number of customers within each age range.

Bars: Show how many customers fall within each age interval.

Insights

Most customers fall within a specific age range.

Businesses can target marketing strategies toward these dominant age groups.

```python
# 4 Histogram (Unit Price Distribution)
plt.figure(figsize=(8,5))
plt.hist(df["unit_price"], bins=20)
plt.title("Unit Price Distribution")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

## Interpretation of your**Chart 4 – Unit Price Distribution (Histogram)**

The above histogram visualizes the distribution of product prices across the platform.

It helps understand the range and frequency of product pricing.

Features

X-axis (Unit Price): Represents product price ranges.

Y-axis (Frequency): Represents the number of products within each price range.

Insights

Most products fall within a certain price range.

Extremely high-priced products appear less frequently.

```python
# 5 Bar Chart (Revenue by Category)
category_revenue = df.groupby("category")["total_order_value"].sum()

category_revenue.plot(kind="bar", figsize=(8,5))
plt.title("Revenue by Category")
plt.ylabel("Total Revenue")
plt.show()
```

## Interpretation of **Chart 5 – Revenue by Category (Bar Chart)**



The above bar chart shows the total revenue generated by each product category.

It compares which categories contribute the most revenue to the platform.

Features

X-axis (Category): Represents different product categories.

Y-axis (Total Revenue): Represents the sum of total order value for each category.

Bars: Show revenue contribution from each category.

Insights

Some categories generate higher revenue than others.

These categories play a major role in the company’s overall sales performance.

```python
# 6 Bar Chart (Purchase Frequency by Country)
country_purchase = df.groupby("country")["purchase_frequency"].mean()
country_purchase.plot(kind="bar")
plt.title("Average Purchase Frequency by Country")
plt.ylabel("Purchase Frequency")
plt.show()
```

## Interpretation of **Chart 6 – Purchase Frequency by Country (Bar Chart)**

The above bar chart visualizes the average purchase frequency of customers from different countries.

Features

X-axis (Country): Represents customer locations.

Y-axis (Average Purchase Frequency): Represents how frequently customers make purchases.

Insights

Customers from certain countries purchase more frequently.

This indicates higher customer engagement in those regions.

```python
# 7 — Pie Chart (Gender Distribution)
df["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Customer Gender Distribution")
plt.ylabel("")
plt.show()
```

## Interpretation of **Chart 7 – Gender Distribution (Pie Chart)**

The above pie chart shows the proportion of male and female customers on the platform.

Features

Pie segments: Represent different gender groups.

Percentage labels: Show the proportion of each gender.

Insights

The chart reveals the gender distribution of the customer base.

Businesses can analyze whether marketing strategies appeal equally to both genders.

```python
# 8 — Violin Plot (Age vs Subscription Status)
plt.figure(figsize=(8,5))
sns.violinplot(x="subscription_status", y="age", data=df)
plt.title("Age Distribution by Subscription Status")
plt.show()
```

## Interpretation of **Chart 8 – Age vs Subscription Status (Violin Plot)**

The above violin plot visualizes the distribution of customer ages across different subscription statuses.

Features

X-axis (Subscription Status): Represents customer subscription categories such as active, paused, or cancelled.

Y-axis (Age): Represents customer age distribution.

Violin shapes: Show the density and spread of ages within each group.

Insights

Different subscription statuses may be associated with different age groups.

This helps identify patterns related to customer behavior.

```python
# 9 — Box Plot (Unit Price by Category)
plt.figure(figsize=(8,5))
sns.boxplot(x="category", y="unit_price", data=df)
plt.title("Unit Price Distribution by Category")
plt.xticks(rotation=45)
plt.show()
```

## Interpretation of **Chart 9 – Unit Price by Category (Box Plot)**

The above box plot visualizes the distribution of product prices within each category.

Features

X-axis (Category): Represents product categories.

Y-axis (Unit Price): Represents product price values.

Box and whiskers: Show median, quartiles, and outliers.

Insights

Some categories contain higher-priced products than others.

The presence of outliers indicates products priced significantly higher than the average.
